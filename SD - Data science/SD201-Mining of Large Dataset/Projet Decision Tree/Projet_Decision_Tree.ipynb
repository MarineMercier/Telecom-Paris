Le but du projet est de coder un decision tree permettant de prédire la valeur de C selon A et B dans le fichier texte situé dans le même dossier.

```python
import numpy as np
import csv
import math
import pandas as pd
```

```python
class Node:
    def __init__(self,l,s):
        self._level=l
        self._statut=s
        self._childs=[]
        self._gini=-1
        self._features=[]
        self._classe=-1
        self._data=[]
    def get_lvl(self):
        return(self._level)
    def set_lvl(self,k):
        self._lvl=k
    def get_statut(self):
        return self._statut
    def set_statut(self,s):
        self._statut=s
    def get_childs(self):
        return self._childs
    def set_childs(self, childs):
        self._childs=childs
    def set_gini(self,k):
        self._gini=k
    def get_gini(self):
        return self._gini
    def set_features(self,f):
        self._features=f
    def get_features(self):
        return self._features 
    def set_classe(self,classe):
        self._classe=classe
    def get_classe(self):
        return self._classe
    def get_data(self):
        return self._data
    def set_data(self,data):
        self._data=data
    level=property(get_lvl,set_lvl)
    gini=property(get_gini,set_gini)
    statut=property(get_statut,set_statut)
    features=property(get_features,set_features)
    classe=property(get_classe,set_classe)
    data=property(get_data,set_data)
    childs=property(get_childs,set_childs)
    
def print_tree_complete(root):
    ToPrint=[root]
    while len(ToPrint)>0: 
        noeud=ToPrint[0]
        print(noeud.statut)
        print("Level "+str(noeud.level))
        if noeud.gini!=-1:
            print("Gini "+str(noeud.gini))
        feat=noeud.features 
        if len(feat)!=0:
            print("Feature " + feat[0] +" ", end="")
            for elmt in feat[1]:
                print(str(elmt)+ " ",end="")
            print()
        if noeud.statut=="leaf" :
            print("Classe "+ str(noeud.classe))
            print(noeud.data)
        ToPrint=ToPrint + (noeud.get_childs())
        ToPrint.pop(0)

def print_tree(root):
    ToPrint=[root]
    while len(ToPrint)>0: 
        noeud=ToPrint[0]
        print(noeud.statut)
        print("Level "+str(noeud.level))
        print("Gini "+str(noeud.gini))
        feat=noeud.features 
        if len(feat)!=0:
            print("Feature " + feat[0] +" ", end="")
            for elmt in feat[1]:
                print(str(elmt)+ " ",end="")
            print()
        ToPrint=ToPrint + (noeud.get_childs())
        ToPrint.pop(0)
        
```

```python
def chercherData(path):
    f=open(path,'r')
    lines=f.readlines()
    M=np.zeros((len(lines)-1,3))
    i=0
    for i in range(1,len(lines)):
        line=lines[i]
        for j in range(3):
            l=line.split()
            M[i][j]=l[j]
        i=i+1
    return (M)

def calculGini(D):
    c0=0
    c1=0
    for i in range(len(D)):
        if D[i][target]==1:
            c1=c1+1
        else: 
            c0=c0+1
    c=c0+c1
    if len(D)!=0:
        return(1-((c0/c)**2+(c1/c)**2))
    else:
        return 1
    
def majority(D,d):
    sum=0
    for i in range(len(D)):
        sum=sum+D[i][target]
    if sum>len(D)/2:
        return 1
    elif sum==len(D)/2:
        return d
    else :
        return 0
    
def calcul_entropy(D):
    sum=0
    if len(D)==0:
        return 0
    for i in range(len(D)):
        sum=sum+D[i][target]
    sum=sum/len(D)
    if sum==0 or sum==1:
        return 0
    entropy=-sum*math.log2(sum)-(1-sum)*math.log2(1-sum)
    return entropy
    
def calcul_information_gain(D,a,v):
    Dfils1=[]
    Dfils2=[]
    for i in range(len(D)):
        if D[i][a] in v:
            Dfils1.append(D[i])
        else:
            Dfils2.append(D[i])
    Info_gain=calcul_entropy(D)-calcul_entropy(Dfils1)*(len(Dfils1)/len(D))-calcul_entropy(Dfils2)*(len(Dfils2)/len(D))
    return Info_gain
        
    
def best_split(D,A):
    best_gain=-math.inf
    best=[]
    for j in range(len(A)):
        values=A[j][1]
        col=A[j][2]
        for i in range(1,len(values)):
            info_gain=calcul_information_gain(D,col,values[:i])
            if (info_gain>best_gain):
                best_gain=info_gain
                best=[A[j],values[:i]]
    return (best[0],best[1])
            
```

```python
def build(D,A,minNum,node,d):
    node.data=D
    node.gini=calculGini(D)
    #Premier cas d'arrêt: toutes les mesures dans la même classe
    c=D[0]
    c=c[target]
    res=0
    for i in range(len(D)):
        if D[i][target]!=c:
            res=1
    if res==0:
        node.statut="leaf"
        node.classe=c
        return
        
    #Deuxième cas d'arrêt: cardinal de D trop petit
    if len(D)<minNum:
        node.statut="leaf"
        node.classe=d
        return
        
    #Troisième cas d'arrêt: A vide 
    if len(A)==0:
        node.statut="leaf"
        node.classe=majority(D,d)
        return
    
    #Recherche de la meilleure partition
    feature,contrainte=best_split(D,A)
    col=feature[2]
    
    #Construction des noeuds fils
    Dfils1=[]
    Dfils2=[]
    for i in range(len(D)):
        if D[i][col] in contrainte:
            Dfils1.append(D[i])
        else:
            Dfils2.append(D[i])
    if (len(Dfils1)==0 or len(Dfils2)==0):
        node.statut='leaf'
        node.classe=majority(D,d)
        
    fils1=Node(node.level+1,"intermediate")
    fils2=Node(node.level+1,"intermediate")

    
    #Mise à jour des informations de node
    node.childs=[fils1,fils2]
    node.features=[feature[0],contrainte]
    node.classe=majority(D,d)
    
    #Suppression du critère utilisé dans A
    index=A .index(feature)
    Abis=list(A)
    Abis.pop(index)
    
    build(Dfils1,Abis,minNum,fils1,d)
    build(Dfils2,Abis,minNum,fils2,d)
    
def buildDecisionTree(path_file,minNum,d):
    data=chercherData(path_file)
    Root=Node(1,"Root")
    build(data,parametres,3,Root,d)
    return Root
```

```python
def gen_errors(root,alpha):
    ToStudy=[root]
    nb_leaf=0
    nb_errors=0
    while len(ToStudy)>0: 
        noeud=ToStudy[0]
        if noeud.statut=="leaf":
            nb_leaf+=1
            c=noeud.classe
            for i in range(len(noeud.data)):
                if noeud.data[i][target]!=c:
                    nb_errors+=1
        ToStudy=ToStudy + (noeud.get_childs())
        ToStudy.pop(0)
    ge=nb_errors+alpha*nb_leaf
    return ge 
    
```

```python
def post_pruning(root,alpha,minNum,d):
    post_prunning_step(root,root,alpha,minNum,d)
    print("generalization error =" +str(gen_errors(root,alpha)))
    return root
    
    
def post_prunning_step(root,node,alpha,minNum,d):
    childs=node.get_childs()
    best_ge=gen_errors(root,alpha)
    if len(childs)!=0:
        post_prunning_step(root,childs[0],alpha,minNum,d)
        post_prunning_step(root,childs[1],alpha,minNum,d)
    if node.statut!='leaf':
        statut_init=node.statut
        node.statut='leaf'
        childs_init=node.childs
        node.set_childs([])
        classe_init=node.classe
        if len(node.data)<minNum:
            node.classe=d
        else :
            node.classe=majority(node.data,d)
        if gen_errors(root,alpha)>=best_ge:
            node.set_childs(childs_init)
            node.statut=statut_init
            node.classe=classe_init
        else: 
            node.features=[]
```

```python
target=2 #Dans le problème étudié la target correspond à l'argument C, dans la 3e colonne
parametres=[['A',[0,1,2,3],0],['B',[0,1,2],1]]
root=buildDecisionTree('test.txt',5,0)
print_tree(root)
print()
post_pruning(root, 0.5,5,0)
print_tree(root)
```
